# Data cleaning script

```python
#Removing duplicates and filter data
```

```python
data = [10, 20, 20, 30, 40, 40, 50]

# Remove duplicates
unique_data = list(set(data))

# Filter values greater than 25
filtered = [x for x in unique_data if x > 25]

print("Unique Data:", unique_data)
print("Filtered Data:", filtered)
```

```python
#User input
```

```python
data = list(map(int, input("Enter numbers separated by space: ").split()))

unique_data = list(set(data))
filtered = [x for x in unique_data if x > 10]

print("Cleaned Data:", filtered)
```
