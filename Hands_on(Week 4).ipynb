# Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\Administrator\\Downloads\\Bank_Loan_Portfolio.csv")
print("\nCSV Data:\n", df.head())
```

```python
# 1. Histogram – Age
plt.figure()
plt.hist(df["Age"], bins=10)
plt.title("Age Distribution")
plt.show()
```

```python
# 2. Histogram – Income
plt.figure()
plt.hist(df["Income"], bins=10)
plt.title("Income Distribution")
plt.show()
```

```python
# 3. Histogram – Credit Score
plt.figure()
plt.hist(df["Credit_Score"], bins=10)
plt.title("Credit Score Distribution")
plt.show()
```

```python
# 4. Line Plot – Loan Amount Trend
plt.figure()
plt.plot(df["Loan_Amount"])
plt.title("Loan Amount Trend")
plt.show()
```

```python
# 5. Scatter – Income vs Loan Amount
plt.figure()
plt.scatter(df["Income"], df["Loan_Amount"])
plt.title("Income vs Loan Amount")
plt.show()
```

```python
# 6. Scatter – Age vs Credit Score
plt.figure()
plt.scatter(df["Age"], df["Credit_Score"])
plt.title("Age vs Credit Score")
plt.show()
```

```python
# 7. Bar – Loan Status Count
plt.figure()
df["Loan_Status"].value_counts().plot(kind="bar")
plt.title("Loan Status Count")
plt.show()
```

```python
# 8. Bar – Approval Status Count
plt.figure()
df["Approval_Status"].value_counts().plot(kind="bar")
plt.title("Approval Status Count")
plt.show()
```

```python
# 9. Pie – Employment Type
plt.figure()
df["Employment_Type"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Employment Type Share")
plt.show()
```

```python
# 10. Pie – Risk Category
plt.figure()
df["Risk_Category"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Risk Category Share")
plt.show()
```

# Seaborn

```python
# 11. Boxplot – Loan Amount by City
plt.figure()
sns.boxplot(x="City", y="Loan_Amount", data=df)
plt.xticks(rotation=45)
plt.title("Loan Amount by City")
plt.show()
```

```python
# 12. Boxplot – Income by Employment Type
plt.figure()
sns.boxplot(x="Employment_Type", y="Income", data=df)
plt.xticks(rotation=45)
plt.title("Income by Employment Type")
plt.show()
```

```python
# 13. Countplot – Region
plt.figure()
sns.countplot(x="Region", data=df)
plt.xticks(rotation=45)
plt.title("Region Count")
plt.show()
```

```python
# 14. Countplot – Default Flag
plt.figure()
sns.countplot(x="Default_Flag", data=df)
plt.title("Default Flag Count")
plt.show()
```

```python
# 15. Violin – Customer Rating vs Loan Status
plt.figure()
sns.violinplot(x="Loan_Status", y="Customer_Rating", data=df)
plt.title("Customer Rating by Loan Status")
plt.show()
```

```python
# 16. Barplot – Avg Interest Rate by Region
plt.figure()
sns.barplot(x="Region", y="Interest_Rate_Percent", data=df)
plt.xticks(rotation=45)
plt.title("Interest Rate by Region")
plt.show()
```

```python
# 17. Barplot – Avg Debt Ratio by Risk Category
plt.figure()
sns.barplot(x="Risk_Category", y="Debt_To_Income_Ratio", data=df)
plt.title("Debt Ratio by Risk Category")
plt.show()
```

```python
# 18. Heatmap – Correlation Matrix
plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 19. Pairplot – Age, Income, Loan Amount
sns.pairplot(df[["Age", "Income", "Loan_Amount"]])
plt.show()
```

```python
# 20. Scatter – Loan Tenure vs Interest Rate
plt.figure()
sns.scatterplot(x="Loan_Tenure_Months", y="Interest_Rate_Percent", data=df)
plt.title("Tenure vs Interest Rate")
plt.show()
```

```python
# 21. Histogram – Customer Rating
plt.figure()
sns.histplot(df["Customer_Rating"], bins=10)
plt.title("Customer Rating Distribution")
plt.show()
```
