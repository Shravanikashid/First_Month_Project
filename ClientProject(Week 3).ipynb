# Data Cleaning and Aggregation

```python
import numpy as np
import pandas as pd
```

```python
# Load Dataset

df = pd.read_csv("C:\\Users\\Administrator\\Downloads\\Bank_Loan_Portfolio.csv")
print("\nCSV Data:\n", df.head())
```

```python
# checking null values of all columns

df.isnull().sum()
```

```python
# Remove missing values

df = df.dropna(subset = ['Income','Employment_Type','Loan_Amount','Credit_Score'])
```

```python
df.isnull().sum()
```

```python
# Removing duplicates

df.drop_duplicates()
```

```python
# Calculate Average
```

```python
avg_income = df["Income"].mean()
print(avg_income)
```

```python
# Groupby category
```

```python
# Average Loan Amount by City

avg_loan_city = df.groupby("City")["Loan_Amount"].mean()
print(avg_loan_city)
```

```python
# Total Income by Region

total_income_region = df.groupby("Region")["Income"].sum()
print(total_income_region)
```

```python
# Average Credit Score by Employment Type

avg_credit_emp = df.groupby("Employment_Type")["Credit_Score"].mean()
print(avg_credit_emp)
```

```python
# Count of Loans by Approval Status

loan_count_approval = df.groupby("Approval_Status")["Loan_ID"].count()
print(loan_count_approval)
```

```python
# Average Income by Risk Category

avg_income_risk = df.groupby("Risk_Category")["Income"].mean()
print(avg_income_risk)
```
