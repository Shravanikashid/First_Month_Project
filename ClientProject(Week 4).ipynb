# Dashboard

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("C:\\Users\\Administrator\\Downloads\\Bank_Loan_Portfolio.csv")
print("\nCSV Data:\n", df.head())
```

```python
# Plots
```

```python
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
sns.scatterplot(x="Income", y="Loan_Amount", data=df)
plt.title("Income vs Loan Amount")

plt.subplot(2,2,2)
sns.histplot(df["Age"], bins=10)
plt.title("Age Distribution")

plt.subplot(2,2,3)
sns.countplot(x="Loan_Status", data=df)
plt.title("Loan Status Count")

plt.subplot(2,2,4)
sns.boxplot(x="Risk_Category", y="Loan_Amount", data=df)
plt.title("Loan Amount by Risk")

plt.tight_layout()
plt.show()
```

```python
# Heatmap

plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

##### I created visualizations using Matplotlib and Seaborn.
##### Scatter plot showed relationship between Income and Loan Amount.
##### Histogram showed most customers are between 25–40 age.
##### Heatmap helped identify correlations between features.
##### Dashboard helps understand loan approval trends.
