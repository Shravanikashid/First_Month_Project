# Numpy

```python
import numpy as np
```

```python
# 1D Array

array_1d = np.array([5, 10, 15, 20, 25])
print("1D Array:", array_1d)
```

```python
# 2D Array

array_2d = np.array([[1, 2, 3],
                     [4, 5, 6]])
print("\n2D Array:\n", array_2d)
```

```python
# 3. Mathematical Operations

print("\nAddition:", array_1d + 2)

print("Multiplication:", array_1d * 3)

print("Subtraction:", array_1d - 1)

print("Division:", array_1d / 2)
```

```python
# 4. Slicing Arrays

print("\nFirst 3 Elements:", array_1d[:3])
print("Last 2 Elements:", array_1d[-2:])
```

```python
# Change Values

array_1d[0] = 100
print("Updated 1D Array:", array_1d)
```

```python
array_2d[0, 1] = 50
print("Updated 2D Array:\n", array_2d)
```

```python
# 5. Broadcasting (Add 10 to all elements)

broadcast_result = array_2d + 10
print("\nBroadcasting Result:\n", broadcast_result)
```

```python
# 6. Statistical Calculations

print("\nMean:", np.mean(array_1d))

print("Sum:", np.sum(array_1d))

print("Maximum:", np.max(array_1d))

print("Minimum:", np.min(array_1d))
```

```python
# Remove missing values

df1 = df1.dropna(subset = ['Income','Employment_Type','Loan_Amount','Credit_Score'])
```

```python
df1.isnull().sum()
```

```python
df1.head()
```

```python
# Average Loan Amount by City

avg_loan_city = df1.groupby("City")["Loan_Amount"].mean()
print(avg_loan_city)
```

```python
# Total Income by Region

total_income_region = df1.groupby("Region")["Income"].sum()
print(total_income_region)
```

```python
# Average Credit Score by Employment Type

avg_credit_emp = df1.groupby("Employment_Type")["Credit_Score"].mean()
print(avg_credit_emp)
```

```python
# Count of Loans by Approval Status

loan_count_approval = df1.groupby("Approval_Status")["Loan_ID"].count()
print(loan_count_approval)
```

```python
# Average Income by Risk Category

avg_income_risk = df1.groupby("Risk_Category")["Income"].mean()
print(avg_income_risk)
```
