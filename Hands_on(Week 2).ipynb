```python
#Lists- ordered ,changeable, allow duplicates
numbers = [1, 2, 3, 4]
numbers.append(5)
```

```python
#Tuples- ordered, unchangeable
t = (1, 2, 3)
```

```python
#Dictionaries- Key-value pairs
student = {"name": "Shravani", "age": 21}
print(student["name"])
```

```python
#Sets- unordered, no duplicates
s = {1, 2, 3, 3}
print(s)   # {1,2,3}
```

```python
#Functions- Reusable block of codes
def greet(name):
    return "Hello " + name
```

```python
#Lambda functions- short, one line anonymous function
square = lambda x: x * x
print(square(4))
```

```python
#Recursion- function calling itself
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)
```

```python
#List Comprehension- Short way to create lists.
squares = [x*x for x in range(5)]
```

## Basic Programs

### 1.Sum of squares

```python
nums = [1, 2, 3, 4]
result = sum([x*x for x in nums])
print(result)
```

### 2.Filtering even numbers

```python
nums = [1,2,3,4,5,6]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)
```

### 3.Function of data transformation

```python
def double_values(lst):
    return [x*2 for x in lst]

print(double_values([1,2,3]))
```
