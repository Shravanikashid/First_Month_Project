```python
#Average Temperature calculator
```

```python
temps = [30, 32, 28, 35, 31]
avg = sum(temps) / len(temps)
print("Average Temperature:", avg)
```

```python
#with user input
```

```python
temps = []
n = int(input("How many days? "))

for i in range(n):
    t = float(input("Enter temperature: "))
    temps.append(t)

avg = sum(temps) / n
print("Average Temperature:", avg)
```
