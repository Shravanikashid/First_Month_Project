```python
#Arithmetic operators

a = 10
b = 3

print("Addition:", a+b)
print("Substraction:", a-b)
print("Multiplication:", a*b)
print("Division:", a/b)
print("Modulus:", a%b)
```

```python
#Comparison operator
x = 10
y = 5

print(x<y)
print(x>y)
print(x==y)
print(x!=y)
```

```python
#Logical operator
a = True
b = False

print(a and b)
print(a or b)
print(not a)
print(not b)
```

```python
#Conditional statement(if-else)
```

```python
#program checks even or odd
num = int(input("Enter a number:"))

if num % 2 == 0:
    print("Number is even")

else:
    print("Number is odd")
```

```python
#program: find greatest of two numbers
num1 = int(input("Enter first number:"))
num2 = int(input("Enter second number:"))

if num1 > num2:
    print("num1 is greater")

else:
    print("num2 is greater")
```
